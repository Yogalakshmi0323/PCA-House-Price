```python
#TO SPECIFICALLY CHECK WHICH COLUMN HAS NULL VALUES AS THERE ARE MORE NUMBER OF COLUMNS
null_values = df.isnull().sum()
columns_with_nulls = null_values[null_values > 0]
print(columns_with_nulls)
```

```python
#FILLING THE NULL VALUES OF THE COLUMNS USING MEDIAN()
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].median())
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())
df
```

```python
#LABEL ENCODING FOR THE COLUMNS TO CONVERT CATEGORICAL INTO NUMERICAL COLUMNS
for col in df.select_dtypes(include = ['object']).columns:
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(df[col].unique())
    df[col] = label_encoder.transform(df[col])
```

```python
df
```

```python
#APPLYING PCA TO OUR DATA BY SPECIFYING NUMBER OF COMPONENTS AS 10 - IT REDUCES THE NUMBER OF COLUMNS TO 10
pca = PCA(n_components = 10)  
dfs = pca.fit_transform(df)
pca_df = pd.DataFrame(data = dfs, columns = ['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5', 'PCA6', 'PCA7', 'PCA8', 'PCA9', 'PCA10'])
#pca_df = pd.DataFrame(data = dfs, columns = [f'PC{i+1}' for i in range(10)]) - #IT AUTO-INCREAMENTS THE COLUMN NAMES
pca_df['SalePrice'] = df['SalePrice']
pca_df
```

```python
#AS THE DATASET IS A REGRESSION DATA WE ARE USING LINEAR REGRESSION, RANDOM FOREST
```

```python
#LINERA REGRESSION
x = pca_df.drop('SalePrice',axis = 1)
y = pca_df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
training_score = linear_reg.score(x_train, y_train)
print("Training score is:",training_score)
print("M / Coefficients are:",linear_reg.coef_)
print("C / Intercept is:",linear_reg.intercept_)
y_pred = linear_reg.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print("MSE:",mse)
rmse = np.sqrt(mse)
print("RMSE:",rmse)
mae = mean_absolute_error(y_test,y_pred)
print("MAE:",mae)
r2 = r2_score(y_test,y_pred)
print("R2 score :",r2)
```

```python
#RANDOM FOREST
rf = RandomForestRegressor(random_state = 42)
rf.fit(x_train,y_train)
rf_training_score = rf.score(x_train,y_train)
print("Training Score in RF:",rf_training_score)
rf_pred = rf.predict(x_test)
r2_score_rf = r2_score(y_test, rf_pred)
print("R2 Score of Random Forest:",r2_score_rf)
```
